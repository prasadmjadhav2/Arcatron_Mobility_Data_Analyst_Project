##### Goals:
- Analyze sales, shipments, and production efficiency.
- Identify top products, regional revenue trends, and shipment efficiency.
- Highlight improvement areas in customer satisfaction and logistics.

##### Basic Insights

```python
# Top-Selling Products

pd.read_sql_query('''
SELECT Product_ID, SUM(Quantity_Sold) AS Total_Quantity
FROM arcatron_mobility_data
GROUP BY Product_ID
ORDER BY Total_Quantity DESC
LIMIT 10;

''', connection)
```

```python
# Total Sales Revenue

pd.read_sql_query('''
SELECT SUM(Sales_Amount) AS Total_Sales_Revenue
FROM arcatron_mobility_data;
''', connection)
```

```python
# Top 5 Products by Revenue

pd.read_sql_query('''
SELECT Product, SUM(Sales_Amount) AS Total_Revenue
FROM arcatron_mobility_data
GROUP BY Product
ORDER BY Total_Revenue DESC
LIMIT 5;

''', connection)
```

```python
# Average Customer Rating by Region

pd.read_sql_query('''
SELECT Region, AVG(Customer_Rating) AS Avg_Rating
FROM arcatron_mobility_data
GROUP BY Region
ORDER BY Avg_Rating DESC;

''', connection)
```

##### Operational Insights

```python
# Shipping Costs by Method

pd.read_sql_query('''
SELECT Shipping_Method, AVG(Logistics_Cost) AS Avg_Shipping_Cost
FROM arcatron_mobility_data
GROUP BY Shipping_Method
ORDER BY Avg_Shipping_Cost DESC;

''', connection)
```

```python
# Average Lead Time by Supplier

pd.read_sql_query('''
SELECT Supplier_ID, AVG(Lead_Time) AS Avg_Lead_Time
FROM arcatron_mobility_data
GROUP BY Supplier_ID
ORDER BY Avg_Lead_Time LIMIT 10;

''', connection)
```

```python
# Production Cost Analysis

pd.read_sql_query('''
SELECT Category, AVG(Production_Cost) AS Avg_Production_Cost
FROM arcatron_mobility_data
GROUP BY Category
ORDER BY Avg_Production_Cost DESC;

''', connection)
```

##### Profitability

```python
# Top 5 Most Profitable Products

pd.read_sql_query('''
SELECT Product, SUM(Profit_Margin) AS Total_Profit
FROM arcatron_mobility_data
GROUP BY Product
ORDER BY Total_Profit DESC
LIMIT 5;

''', connection)
```

```python
# Profitability by Region

pd.read_sql_query('''
SELECT Region, SUM(Profit_Margin) AS Total_Profit
FROM arcatron_mobility_data
GROUP BY Region
ORDER BY Total_Profit DESC;

''', connection)
```

##### Customer Satisfaction

```python
# Delivery Satisfaction by Shipping Method

pd.read_sql_query('''
SELECT Shipping_Method, AVG(Delivery_Satisfaction) AS Avg_Satisfaction
FROM arcatron_mobility_data
GROUP BY Shipping_Method
ORDER BY Avg_Satisfaction DESC;

''', connection)
```

```python
# Most Common Customer Feedback

pd.read_sql_query('''
SELECT Customer_Feedback, COUNT(*) AS Feedback_Count
FROM arcatron_mobility_data
GROUP BY Customer_Feedback
ORDER BY Feedback_Count DESC
LIMIT 5;

''', connection)
```

##### Inventory Insights

```python
# Products Below Reorder Level

pd.read_sql_query('''
SELECT Product_Code, Stock_Quantity, Reorder_Level
FROM arcatron_mobility_data
WHERE Stock_Quantity < Reorder_Level LIMIT 10;

''', connection)
```

```python
# Stock Valuation by Warehouse

pd.read_sql_query('''
SELECT Warehouse_ID, SUM(Stock_Valuation) AS Total_Stock_Value
FROM arcatron_mobility_data
GROUP BY Warehouse_ID
ORDER BY Total_Stock_Value DESC LIMIT 10;

''', connection)
```

```python
# Advanced Queries
```

```python
# Delay in Delivery Analysis

pd.read_sql_query('''
SELECT 
    Shipment_ID, 
    DATEDIFF(Delivery_Date, Shipping_Date) AS Delivery_Days 
FROM arcatron_mobility_data
WHERE DATEDIFF(Delivery_Date, Shipping_Date) > 1;

''', connection)
```

```python
# Marketing Campaign Effectiveness

pd.read_sql_query('''
SELECT Marketing_Campaign, AVG(Profit_Margin) AS Avg_Profit
FROM arcatron_mobility_data
GROUP BY Marketing_Campaign
ORDER BY Avg_Profit DESC;

''', connection)
```

```python
# Identify Top Suppliers with Low Lead Time and High Profitability

pd.read_sql_query('''
SELECT 
    Supplier_ID, 
    AVG(Lead_Time) AS Avg_Lead_Time, 
    SUM(Profit_Margin) AS Total_Profit
FROM arcatron_mobility_data
GROUP BY Supplier_ID
HAVING AVG(Lead_Time) < 5 AND Total_Profit > 10000
ORDER BY Total_Profit DESC;

''', connection)
```

```python
# Customer Retention Analysis
# Identify customers who placed multiple orders in a given period.

pd.read_sql_query('''
SELECT 
    Customer_ID, 
    COUNT(Order_ID) AS Total_Orders, 
    SUM(Sales_Amount) AS Total_Spent
FROM arcatron_mobility_data
WHERE Order_Date BETWEEN '2024-01-01' AND '2024-12-31'
GROUP BY Customer_ID
HAVING Total_Orders > 3
ORDER BY Total_Spent DESC;

''', connection)
```

```python
# Shipping Method Performance with Customer Satisfaction

pd.read_sql_query('''
SELECT 
    Shipping_Method, 
    COUNT(Shipment_ID) AS Total_Shipments,
    AVG(Delivery_Satisfaction) AS Avg_Customer_Satisfaction,
    AVG(Logistics_Cost) AS Avg_Cost_Per_Shipment
FROM arcatron_mobility_data
GROUP BY Shipping_Method
HAVING Avg_Customer_Satisfaction > 4
ORDER BY Avg_Cost_Per_Shipment ASC;

''', connection)
```

```python
# Product Demand Forecasting
# Identify products with consistent sales growth over time.

pd.read_sql_query('''
SELECT 
    Product_ID, 
    MONTH(Order_Date) AS Month, 
    SUM(Quantity_Sold) AS Total_Sales
FROM arcatron_mobility_data
GROUP BY Product_ID, Month
ORDER BY Product_ID, Month LIMIT 10;

''', connection)
```

```python
# Correlation Between Defect Flags and Lead Time

pd.read_sql_query('''
SELECT 
    Lead_Time, 
    COUNT(Defect_Flag) AS Total_Defects
FROM arcatron_mobility_data
WHERE Defect_Flag = 'Yes'
GROUP BY Lead_Time
ORDER BY Lead_Time ASC;

''', connection)
```

```python
# Profitability by Region and Category

pd.read_sql_query('''
SELECT 
    Region, 
    Category, 
    SUM(Sales_Amount) AS Total_Sales, 
    SUM(Profit_Margin) AS Total_Profit
FROM arcatron_mobility_data
GROUP BY Region, Category
ORDER BY Total_Profit DESC;

''', connection)
```

```python
# Seasonal Sales Trends
# Analyze how sales vary by season (grouped by quarters).

pd.read_sql_query('''
SELECT 
    QUARTER(Order_Date) AS Quarter, 
    Category, 
    SUM(Sales_Amount) AS Total_Sales
FROM arcatron_mobility_data
GROUP BY Quarter, Category
ORDER BY Quarter, Total_Sales DESC;

''', connection)
```

```python
# Underperforming Products
# Identify products with high returns and low customer ratings.

pd.read_sql_query('''
SELECT 
    Product, 
    COUNT(Return_Flag) AS Total_Returns, 
    AVG(Customer_Rating) AS Avg_Rating
FROM arcatron_mobility_data
WHERE Return_Flag = 'Yes'
GROUP BY Product
HAVING Avg_Rating < 3
ORDER BY Total_Returns DESC;

''', connection)
```

```python
# High-Cost Orders with Low Satisfaction

pd.read_sql_query('''
SELECT 
    Order_ID, 
    Customer_ID, 
    Sales_Amount, 
    Delivery_Satisfaction
FROM arcatron_mobility_data
WHERE Sales_Amount > 1000 AND Delivery_Satisfaction < 3
ORDER BY Sales_Amount DESC LIMIT 10;

''', connection)
```

```python
# Customer Segmentation by Spending

pd.read_sql_query('''
SELECT 
    Customer_ID,
    CASE 
        WHEN SUM(Sales_Amount) < 500 THEN 'Low Spender'
        WHEN SUM(Sales_Amount) BETWEEN 500 AND 2000 THEN 'Mid Spender'
        ELSE 'High Spender'
    END AS Spending_Category,
    SUM(Sales_Amount) AS Total_Spent
FROM arcatron_mobility_data
GROUP BY Customer_ID
ORDER BY Total_Spent DESC LIMIT 10;

''', connection)
```

```python
# Operational Cost vs. Profit Margin Analysis

pd.read_sql_query('''
SELECT 
    Region,
    AVG(Operational_Cost) AS Avg_Operational_Cost,
    AVG(Profit_Margin) AS Avg_Profit_Margin,
    (AVG(Profit_Margin) / AVG(Operational_Cost)) * 100 AS Profitability_Ratio
FROM arcatron_mobility_data
GROUP BY Region
ORDER BY Profitability_Ratio DESC;

''', connection)
```

```python
# Impact of Warranty Period on Returns

pd.read_sql_query('''
SELECT 
    Warranty_Period, 
    COUNT(Return_Flag) AS Total_Returns
FROM arcatron_mobility_data
WHERE Return_Flag = 'Yes'
GROUP BY Warranty_Period
ORDER BY Warranty_Period ASC;

''', connection)
```

```python
# Marketing Campaign Effectiveness by Region

pd.read_sql_query('''
SELECT 
    Marketing_Campaign,
    Region,
    SUM(Sales_Amount) AS Total_Sales,
    AVG(Profit_Margin) AS Avg_Profit_Margin
FROM arcatron_mobility_data
GROUP BY Marketing_Campaign, Region
ORDER BY Total_Sales DESC, Avg_Profit_Margin DESC;

''', connection)
```

```python
# Delivery Delays by Region

pd.read_sql_query('''
SELECT 
    Region, 
    COUNT(*) AS Total_Delays,
    AVG(DATEDIFF(Delivery_Date, Shipping_Date)) AS Avg_Delay_Days
FROM arcatron_mobility_data
WHERE DATEDIFF(Delivery_Date, Shipping_Date) > 0
GROUP BY Region
ORDER BY Total_Delays DESC;

''', connection)
```

```python
# Outliers in Sales Data
# Detect unusually high or low sales.

pd.read_sql_query('''
SELECT 
    Order_ID, 
    Product, 
    Sales_Amount
FROM arcatron_mobility_data
WHERE Sales_Amount > (SELECT AVG(Sales_Amount) + 2 * STDDEV(Sales_Amount) FROM arcatron_mobility_data)
   OR Sales_Amount < (SELECT AVG(Sales_Amount) - 2 * STDDEV(Sales_Amount) FROM arcatron_mobility_data)
ORDER BY Sales_Amount DESC;

''', connection)
```

```python
# Find Customers Who Spend More Than the Average Spending

pd.read_sql_query('''
SELECT Customer_ID, Total_Spent
FROM (
    SELECT 
        Customer_ID,
        SUM(Sales_Amount) AS Total_Spent
    FROM arcatron_mobility_data
    GROUP BY Customer_ID
) AS Customer_Spending
WHERE Total_Spent > (
    SELECT AVG(Total_Spent) 
    FROM (
        SELECT 
            Customer_ID,
            SUM(Sales_Amount) AS Total_Spent
        FROM arcatron_mobility_data
        GROUP BY Customer_ID
    ) AS Avg_Customer_Spending
)
ORDER BY Total_Spent DESC LIMIT 10;

''', connection)
```

```python
# Best Performing Product in Each Category

pd.read_sql_query('''
SELECT Product, Category, Total_Sales
FROM (
    SELECT 
        Product, 
        Category, 
        SUM(Sales_Amount) AS Total_Sales,
        RANK() OVER (PARTITION BY Category ORDER BY SUM(Sales_Amount) DESC) AS Rank
    FROM arcatron_mobility_data
    GROUP BY Product, Category
) AS Ranked_Products
WHERE Rank = 1;

''', connection)
```

```python
# Identify Suppliers Contributing to Top 10% of Profits

pd.read_sql_query('''
SELECT Supplier_ID, Total_Profit
FROM (
    SELECT 
        Supplier_ID,
        SUM(Profit_Margin) AS Total_Profit
    FROM arcatron_mobility_data
    GROUP BY Supplier_ID
) AS Supplier_Profits
WHERE Total_Profit >= (
    SELECT PERCENTILE_CONT(0.9) WITHIN GROUP (ORDER BY Total_Profit)
    FROM (
        SELECT 
            Supplier_ID,
            SUM(Profit_Margin) AS Total_Profit
        FROM arcatron_mobility_data
        GROUP BY Supplier_ID
    ) AS Supplier_Profits_Subquery
);

''', connection)
```

```python
# Find Customers Who Ordered Products from All Categories

pd.read_sql_query('''
SELECT Customer_ID
FROM arcatron_mobility_data
GROUP BY Customer_ID
HAVING COUNT(DISTINCT Category) = (
    SELECT COUNT(DISTINCT Category)
    FROM arcatron_mobility_data
) LIMIT 10;

''', connection)
```

```python
# Detect Products with Below-Average Ratings in Their Category

pd.read_sql_query('''
SELECT Product, Category, Avg_Rating
FROM (
    SELECT 
        Product, 
        Category, 
        AVG(Customer_Rating) AS Avg_Rating
    FROM arcatron_mobility_data
    GROUP BY Product, Category
) AS Product_Ratings
WHERE Avg_Rating < (
    SELECT AVG(Customer_Rating)
    FROM arcatron_mobility_data
    WHERE Category = Product_Ratings.Category
);

''', connection)
```

```python
# Regions with Consistently High Sales Across All Quarters

pd.read_sql_query('''
SELECT Region
FROM (
    SELECT 
        Region, 
        QUARTER(Order_Date) AS Quarter, 
        SUM(Sales_Amount) AS Total_Sales
    FROM arcatron_mobility_data
    GROUP BY Region, Quarter
) AS Quarterly_Sales
GROUP BY Region
HAVING MIN(Total_Sales) > (
    SELECT AVG(SUM(Sales_Amount))
    FROM arcatron_mobility_data
    GROUP BY QUARTER(Order_Date)
);

''', connection)
```

```python
# Identify Underperforming Shipping Methods

pd.read_sql_query('''
SELECT Shipping_Method
FROM (
    SELECT 
        Shipping_Method, 
        AVG(DATEDIFF(Order_Date, Delivery_Date)) AS Avg_Delivery_Time 
    FROM arcatron_mobility_data
    GROUP BY Shipping_Method
) AS Shipping_Method_Averages
WHERE Avg_Delivery_Time > (
    SELECT AVG(Avg_Delivery_Time)
    FROM (
        SELECT 
            AVG(DATEDIFF(Order_Date, Delivery_Date)) AS Avg_Delivery_Time
        FROM arcatron_mobility_data
        GROUP BY Shipping_Method
    ) AS All_Shipping_Method_Averages
);

''', connection)
```

```python
# Find Products Returned More Frequently Than Average

pd.read_sql_query('''
SELECT Product, Total_Returns
FROM (
    SELECT 
        Product, 
        COUNT(Return_Flag) AS Total_Returns
    FROM arcatron_mobility_data
    WHERE Return_Flag = 'Yes'
    GROUP BY Product
) AS Product_Returns
WHERE Total_Returns > (
    SELECT AVG(COUNT(Return_Flag))
    FROM arcatron_mobility_data
    WHERE Return_Flag = 'Yes'
    GROUP BY Product
);

''', connection)
```

```python
# Average Sales by Top Regions Based on Total Profit

pd.read_sql_query('''
SELECT Region, Avg_Sales
FROM (
    SELECT 
        Region, 
        AVG(Sales_Amount) AS Avg_Sales, 
        RANK() OVER (ORDER BY SUM(Profit_Margin) DESC) AS Profit_Rank
    FROM arcatron_mobility_data
    GROUP BY Region
) AS Region_Stats
WHERE Profit_Rank <= 3;

''', connection)
```

```python
# Find the Maximum Sales Difference Between Categories

pd.read_sql_query('''
SELECT MAX(Sales_Difference) AS Max_Sales_Difference
FROM (
    SELECT 
        Category, 
        MAX(Sales_Amount) - MIN(Sales_Amount) AS Sales_Difference
    FROM arcatron_mobility_data
    GROUP BY Category
) AS Category_Sales_Diff;

''', connection)
```

```python
# Correlate Defect Flags with Supplier Performance

pd.read_sql_query('''
SELECT Supplier_ID, Defect_Rate
FROM (
    SELECT 
        Supplier_ID, 
        SUM(CASE WHEN Defect_Flag = 'Yes' THEN 1 ELSE 0 END) / COUNT(*) AS Defect_Rate
    FROM arcatron_mobility_data
    GROUP BY Supplier_ID
) AS Supplier_Defects
WHERE Defect_Rate > (
    SELECT AVG(SUM(CASE WHEN Defect_Flag = 'Yes' THEN 1 ELSE 0 END) / COUNT(*))
    FROM arcatron_mobility_data
    GROUP BY Supplier_ID
);

''', connection)
```

```python
# Calculate Customer Lifetime Value (CLV)

pd.read_sql_query('''
SELECT Customer_ID, Lifetime_Value
FROM (
    SELECT 
        Customer_ID, 
        SUM(Sales_Amount) AS Lifetime_Value
    FROM arcatron_mobility_data
    GROUP BY Customer_ID
) AS Customer_Values
WHERE Lifetime_Value > (
    SELECT AVG(SUM(Sales_Amount))
    FROM arcatron_mobility_data
    GROUP BY Customer_ID
);

''', connection)
```

```python
# Product Popularity Index

pd.read_sql_query('''
SELECT Product, Popularity_Index
FROM (
    SELECT 
        Product, 
        (SUM(Quantity_Sold) * AVG(Customer_Rating)) AS Popularity_Index
    FROM arcatron_mobility_data
    GROUP BY Product
) AS Product_Popularity
ORDER BY Popularity_Index DESC;

''', connection)
```
